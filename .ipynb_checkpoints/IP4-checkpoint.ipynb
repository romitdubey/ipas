1. Write a program for intensity transformation of an image using following transformations:
*  Image Negative and identity transformation
*  Log Transformation and Inverse Log Transformation
*  Power Law (use different values of gamma as shown in graph: refer notes)
*  nth root and nth power transformation

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

def load_image(image_path):
    image = tf.io.read_file(image_path)
    image = tf.image.decode_image(image, channels=3)
    image = tf.image.rgb_to_grayscale(image)
    return np.array(image, dtype=np.uint8)

image_path = "Lenna.jpeg"
gray_image = load_image(image_path)

plt.imshow(gray_image, cmap='gray')
plt.show()
```

```python
def image_negative(image):
    return 255 - image

def log_transformation(image):
    c = 255 / np.log(1 + np.max(image))
    return (c * np.log(1 + image + 1e-5)).astype(np.uint8)

def inverse_log_transformation(image):
    c = 255 / np.log(1 + np.max(image))
    return (np.exp(image / c) - 1).astype(np.uint8)

def power_law_transformation(image, gamma):
    c = 255 / (255 ** gamma)
    return (c * (image ** gamma)).astype(np.uint8)

def nth_root_transformation(image, n):
    return ((image / 255) ** (1 / n) * 255).astype(np.uint8)

def nth_power_transformation(image, n):
    return ((image / 255) ** n * 255).astype(np.uint8)

neg_image = image_negative(gray_image)
log_image = log_transformation(gray_image)
inv_log_image = inverse_log_transformation(gray_image)
power_image = power_law_transformation(gray_image, 0.5)
root_image = nth_root_transformation(gray_image, 2)
power_n_image = nth_power_transformation(gray_image, 2)

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
titles = ["Negative", "Log", "Inverse Log", "Power Law (γ=0.5)", "Nth Root", "Nth Power"]
images = [neg_image, log_image, inv_log_image, power_image, root_image, power_n_image]

for ax, img, title in zip(axes.flatten(), images, titles):
    ax.imshow(img, cmap='gray')
    ax.set_title(title)
plt.show()
```

2. Implement a function to perform bit plane slicing on the 8-bit grayscale image of size M*N.
*  The function should extract individual bit planes ranging from 0 to 7.
*  Visualize the original grayscale image along with its individual bit planes.

```python
def bit_plane_slicing(image):
    bit_planes = [(image & (1 << i)) >> i for i in range(8)]
    fig, axes = plt.subplots(1, 9, figsize=(20, 5))
    axes[0].imshow(image, cmap='gray')
    axes[0].set_title('Original Image')
    for i in range(8):
        axes[i+1].imshow(bit_planes[i] * 255, cmap='gray')
        axes[i+1].set_title(f'Bit Plane {i}')
    plt.show()

bit_plane_slicing(gray_image)
```

3. Hide the message within the least significant bit (LSB) of the pixel values in the image.
*  Take an image of size M*N
*  Take a secret message of size MN (eg. 10101000011....................MN)
*  embed the message in the last LSB of input image.
*  Show input and embedded image.
*  Extract message from embedded image.
*  Plot histogram of input image, embedded image and image obtained after message extraction.

```python
import numpy as np
import matplotlib.pyplot as plt

def embed_message(image, message):
    image = image.astype(np.uint8)
    total_pixels = image.shape[0] * image.shape[1]
    message = (message * (total_pixels // len(message) + 1))[:total_pixels]

    message_array = np.array([int(bit) for bit in message], dtype=np.uint8).reshape(image.shape[:2])
    embedded_image = (image & 0b11111110) | message_array[..., np.newaxis]

    return embedded_image

def extract_message(embedded_image):
    return (embedded_image & 1).flatten()

def plot_histograms(original, embedded, extracted):
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))

    axes[0].hist(original.ravel(), bins=256, color='black')
    axes[0].set_title("Original Image Histogram")

    axes[1].hist(embedded.ravel(), bins=256, color='red')
    axes[1].set_title("Embedded Image Histogram")

    axes[2].hist(extracted.ravel(), bins=256, color='blue')
    axes[2].set_title("Extracted Message Histogram")

    plt.show()

message = "1010100011"
total_pixels = gray_image.shape[0] * gray_image.shape[1]
message = (message * (total_pixels // len(message) + 1))[:total_pixels]

embedded_image = embed_message(gray_image, message)

extracted_message = extract_message(embedded_image)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(embedded_image, cmap='gray')
plt.title("Embedded Image")

plt.show()

plot_histograms(gray_image, embedded_image, np.array(extracted_message).reshape(gray_image.shape[:2]))
```

4. Implement histogram equalization on the grayscale image
*  Display both the original grayscale image and the histogram-equalized image
*  side by side for comparison.
*  Display both the histogram of original grayscale image and the histogram of equalized image side by side for comparison.

```python
import cv2
def histogram_equalization(image):
    equalized = cv2.equalizeHist(image)
    fig, axes = plt.subplots(2, 2, figsize=(10, 8))

    axes[0, 0].imshow(image, cmap='gray')
    axes[0, 0].set_title("Original Image")
    axes[0, 1].imshow(equalized, cmap='gray')
    axes[0, 1].set_title("Equalized Image")

    axes[1, 0].hist(image.ravel(), bins=256, color='black')
    axes[1, 0].set_title("Histogram of Original Image")
    axes[1, 1].hist(equalized.ravel(), bins=256, color='blue')
    axes[1, 1].set_title("Histogram of Equalized Image")

    plt.show()
    return equalized

histogram_equalization(gray_image)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def compute_histogram(image):
    hist = np.zeros(256, dtype=int)
    for pixel in image.flatten():
        hist[pixel] += 1
    return hist

def histogram_equalization(image):
    # Step 1: Compute histogram
    hist = compute_histogram(image)

    # Step 2: Compute CDF (Cumulative Distribution Function)
    cdf = np.cumsum(hist)
    cdf_min = cdf[np.nonzero(cdf)][0]  # first non-zero value in cdf
    total_pixels = image.size

    # Step 3: Normalize the CDF
    cdf_normalized = (cdf - cdf_min) * 255 / (total_pixels - cdf_min)
    cdf_normalized = cdf_normalized.astype('uint8')  # mapping values

    # Step 4: Map original image pixels to equalized pixels
    equalized_image = cdf_normalized[image]

    return equalized_image, hist, compute_histogram(equalized_image)

# ------------------- USAGE -------------------

# Example grayscale image (use your own image here)
# Use: gray_image = cv2.imread('image.png', 0)
# or any grayscale array
gray_image = np.random.randint(0, 256, (256, 256), dtype=np.uint8)  # Dummy example

equalized_img, original_hist, equalized_hist = histogram_equalization(gray_image)

# Display images side by side
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equalized_img, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')
plt.show()

# Display histograms side by side
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(original_hist, color='black')
plt.title('Original Histogram')

plt.subplot(1, 2, 2)
plt.plot(equalized_hist, color='black')
plt.title('Equalized Histogram')
plt.show()
```
