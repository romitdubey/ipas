2.Write a program to perform basic arithmetic operations on two images:
*  Addition
*  Subtraction
*  Multiplication
*  Averaging

```python
img1 = tf.keras.utils.load_img('Lenna.jpeg')
img1_array = np.array(img1, dtype="uint8")
img2 = tf.keras.utils.load_img('cpeppers.jpg')
img2_array = np.array(img2, dtype="uint8")
```

```python
fig, axs = plt.subplots(1, 2)
axs[0].imshow(img1)
axs[0].set_title("Image 1")
axs[1].imshow(img2)
axs[1].set_title("Image 2")
axs[0].axis('off')
axs[1].axis('off')

plt.show()
```

# Addition

```python
# checking the shape of both the images

print(f"Image 1 shape: {img1_array.shape}")
print(f"Image 2 shape : {img2_array.shape}")
```

```python
img1_array = tf.image.resize(img1_array, [128,128])
img2_array = tf.image.resize(img2_array, [128,128])
```

```python
print(f"Image 1 shape: {img1_array.shape}")
print(f"Image 2 shape : {img2_array.shape}")
```

```python
added_image = np.zeros_like(img1_array)

for i in range(img1_array.shape[0]):
  for j in range(img1_array.shape[1]):
    for k in range(img1_array.shape[2]):
      added_image[i, j, k] = np.clip(img1_array[i, j, k] + img2_array[i, j, k], 0, 255)

plt.imshow(added_image.astype('uint8'))
plt.title("Added Image")
plt.axis('off')
plt.show()
```

## Substraction

```python
substracted_image = np.zeros_like(img1_array)

for i in range(img1_array.shape[0]):
  for j in range(img1_array.shape[1]):
    for k in range(img1_array.shape[2]):
      substracted_image[i, j, k] = np.clip(img1_array[i, j, k] - img2_array[i, j, k], 0, 255)

plt.imshow(substracted_image.astype('uint8'))
plt.title("Substracted Image")
plt.axis('off')
plt.show()
```

## Multiplication

```python
multiplied_image = np.zeros_like(img1_array)

for i in range(img1_array.shape[0]):
  for j in range(img1_array.shape[1]):
    for k in range(img1_array.shape[2]):
      multiplied_image[i, j, k] = np.clip(img1_array[i, j, k] * img2_array[i, j, k], 0, 255)

plt.imshow(multiplied_image.astype('uint8'))
plt.title("Multiplied Image")
plt.axis('off')
plt.show()
```

## Average

```python
averaged_image = np.zeros_like(img1_array)

for i in range(img1_array.shape[0]):
  for j in range(img1_array.shape[1]):
    for k in range(img1_array.shape[2]):
      averaged_image[i, j, k] = np.clip((img1_array[i, j, k] + img2_array[i, j, k]) / 2, 0, 255)

plt.imshow(averaged_image.astype('uint8'))
plt.title("Averaged Image")
plt.axis('off')
plt.show()
```

3.  Take an image of size 512 * 512, sub-sampled down it to size 256 * 256, 128 * 128, 64 * 64, 32 * 32
pixels. The number of allowable gray levels was kept at 256. Sub-sampling
is achieved by deleting every other row and column.




```python
colour_image = tf.keras.utils.load_img('Lenna.jpeg')
colour_image_array = tf.keras.utils.img_to_array(colour_image)
gray_image = tf.image.rgb_to_grayscale(colour_image_array)
gray_image_array = np.array(gray_image, dtype="uint8")

subsample_256x256 = gray_image_array[::2, ::2]
subsample_128x128 = gray_image_array[::4, ::4]
subsample_64x64 = gray_image_array[::8, ::8]
subsample_32x32 = gray_image_array[::16, ::16]


fig, axs = plt.subplots(2, 3, figsize=(15, 10))

axs[0, 0].imshow(gray_image_array, cmap='gray')
axs[0, 0].set_title("Original Image")
axs[0, 0].axis('off')
axs[0,1].imshow(subsample_256x256, cmap='gray')
axs[0,1].set_title("Sub-Sampled_256x256")
axs[0,1].axis('off')
axs[0,2].imshow(subsample_128x128, cmap='gray')
axs[0,2].set_title("Sub-Sampled_128x1258")
axs[0,2].axis('off')
axs[1,0].imshow(subsample_64x64, cmap='gray')
axs[1,0].set_title("Sub-Sampled_64x64")
axs[1,0].axis('off')
axs[1,1].imshow(subsample_32x32, cmap='gray')
axs[1,1].set_title("Sub-Sampled_32x32")
axs[1,1].axis('off')
axs[1,2].axis('off')

plt.show()
```

4. Apply Re-sampling, on images obtained in Q. 3 by row and column duplication
*  256 * 256 Image resampled into 512 * 512 pixels by row and column duplication.
*  128 * 128 Image resampled into 512 * 512 pixels by row and column duplication.  
*  64 * 64  Image resampled into 512 * 512 pixels by row and column duplication.  
*  32 * 32  Image resampled into 512 * 512 pixels by row and column duplication.

```python
import numpy as np

def resample_image(image, target_size):

    height, width = image.shape[:2]
    target_height, target_width = target_size
    row_factor = target_height // height
    col_factor = target_width // width

    resampled_image = np.zeros((target_height, target_width), dtype=image.dtype)

    for i in range(height):
        for j in range(width):

            for x in range(row_factor):
                for y in range(col_factor):
                    resampled_image[i * row_factor + x, j * col_factor + y] = image[i, j]

    return resampled_image

resampled_256 = resample_image(subsample_256x256, (512,512))
resampled_128 = resample_image(subsample_128x128, (512,512))
resampled_64 = resample_image(subsample_64x64, (512,512))
resampled_32 = resample_image(subsample_32x32, (512,512))

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

axs[0, 0].imshow(resampled_256, cmap='gray')
axs[0, 0].set_title("Resampled 256x256")
axs[0, 0].axis('off')

axs[0, 1].imshow(resampled_128, cmap='gray')
axs[0, 1].set_title("Resampled 128x128")
axs[0, 1].axis('off')

axs[1, 0].imshow(resampled_64, cmap='gray')
axs[1, 0].set_title("Resampled 64x64")
axs[1, 0].axis('off')

axs[1, 1].imshow(resampled_32, cmap='gray')
axs[1, 1].set_title("Resampled 32x32")
axs[1, 1].axis('off')

plt.show()
```
